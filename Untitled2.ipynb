```python
import os
import gdown
import subprocess
from concurrent.futures import ThreadPoolExecutor

drive_links = [
    "https://drive.google.com/file/d/1D4_ZpidCEVJv5TiiZAgcXp7_155Sz3A8/view?usp=drive_link",
    "https://drive.google.com/file/d/1UrMQeJcFb_eWcozhlYv4dThE3A9f0GIg/view?usp=drive_link",
    "https://drive.google.com/file/d/1DkHc_SGBtbQBHERZZ6L0yehM_qiCcCyW/view?usp=drive_link",
    "https://drive.google.com/file/d/1ld15uE74yoj3W6kWs7dBKsHOz0rahRfA/view?usp=drive_link",
    "https://drive.google.com/file/d/1JQYu6E7kzCQQkJpGCXZoskd9sowkp5SS/view?usp=drive_link",
    "https://drive.google.com/file/d/1PB9-iDUm-L39zAz_MG2krgyLexn2yeW7/view?usp=drive_link"
]

durations = [
    "00:08:48",
    "00:05:54",
    "00:07:23",
    "00:06:56",
    "00:07:20",
    "00:06:02"
]

severities = [
    "moderate",
    "mild",
    "marked",
    "moderate",
    "moderate",
    "moderate"
]

output_base_folder = "autism"
os.makedirs(output_base_folder, exist_ok=True)

for severity in set(severities):
    os.makedirs(os.path.join(output_base_folder, severity), exist_ok=True)

def download_video_from_drive(drive_link, output_path):
    file_id = drive_link.split('/d/')[1].split('/view')[0]
    gdown.download(f"https://drive.google.com/uc?export=download&id={file_id}", output_path, quiet=False)

def video_to_frames_ffmpeg(video_path, severity, duration, frame_count_offset=0, frame_rate=1):
    output_folder = os.path.join(output_base_folder, severity)

    output_pattern = os.path.join(output_folder, f"frame_%04d.jpg")
    command = f"ffmpeg -i {video_path} -t {duration} -vf 'fps={frame_rate}' -start_number {frame_count_offset} {output_pattern}"
    subprocess.run(command, shell=True)

def process_video(i, link, duration, severity):
    video_path = f"/content/video_{i}.mp4"
    print(f"Downloading Video {i+1} ({severity})...")
    download_video_from_drive(link, video_path)

    print(f"Extracting Frames from Video {i+1} ({severity}) for duration {duration}...")

    frame_count_offset = 0
    for idx in range(i):
        if severities[idx] == severity:
            prev_duration = durations[idx]
            h, m, s = map(int, prev_duration.split(':'))
            seconds = h * 3600 + m * 60 + s
            frame_count_offset += seconds

    video_to_frames_ffmpeg(video_path, severity, duration, frame_count_offset, frame_rate=1)

    print(f"Completed processing Video {i+1} ({severity})")

with ThreadPoolExecutor(max_workers=6) as executor:
    executor.map(process_video, range(len(drive_links)), drive_links, durations, severities)

print("All videos processed successfully!")
```

```python
import os
import cv2

VIDEO_DIR = "avspeech_clips"
OUTPUT_DIR = os.path.join("autism", "not_present")  # All frames go here
os.makedirs(OUTPUT_DIR, exist_ok=True)

def extract_frames(video_path, fps=1):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"❌ Failed to open {video_path}")
        return

    video_name = os.path.splitext(os.path.basename(video_path))[0]
    frame_rate = cap.get(cv2.CAP_PROP_FPS)
    interval = int(frame_rate / fps)

    frame_count = 0
    saved_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if frame_count % interval == 0:
            # Prefix frame filename with video name to avoid collisions
            filename = os.path.join(OUTPUT_DIR, f"{video_name}_frame_{saved_count:04d}.jpg")
            cv2.imwrite(filename, frame)
            saved_count += 1

        frame_count += 1

    cap.release()
    print(f"✅ Extracted {saved_count} frames from {video_path}")

# Process all videos in the folder
for file in os.listdir(VIDEO_DIR):
    if file.endswith(".mp4"):
        extract_frames(os.path.join(VIDEO_DIR, file))
```

```python
import os
import glob
from PIL import Image
import torch
from torch.utils.data import Dataset
from torchvision import transforms

class VideoClipDataset(Dataset):
    def __init__(self, root_dir, clip_len=16, transform=None):
        self.root_dir = root_dir
        self.clip_len = clip_len
        self.transform = transform

        self.class_map = {'not_present': 0, 'mild': 1, 'moderate': 2, 'marked': 3, 'severe': 4}
        self.clips = []
        self.labels = []

        for label_name in os.listdir(root_dir):
            label_path = os.path.join(root_dir, label_name)
            frame_files = sorted(glob.glob(os.path.join(label_path, "*.jpg")))

            for i in range(0, len(frame_files) - clip_len + 1, clip_len):
                clip = frame_files[i:i+clip_len]
                self.clips.append(clip)
                self.labels.append(self.class_map[label_name])

    def __len__(self):
        return len(self.clips)

    def __getitem__(self, idx):
        clip_paths = self.clips[idx]
        clip = [self.transform(Image.open(frame).convert("RGB")) for frame in clip_paths]
        clip = torch.stack(clip, dim=1)  # (C, T, H, W)
        label = self.labels[idx]
        return clip, label

# Transform
transform = transforms.Compose([
    transforms.Resize((112, 112)),
    transforms.ToTensor()
])

# Create dataset
dataset_root = "autism"  # Assuming structure like: autism/not_present/*.jpg, autism/mild/*.jpg, etc.
dataset = VideoClipDataset(root_dir=dataset_root, clip_len=16, transform=transform)

# Print total samples
print(f"Total number of samples: {len(dataset)}")
```

```python
import torchvision.models.video as video_models
import torch.nn as nn
import torch

class Hybrid3DCNN(nn.Module):
    def __init__(self, num_classes=5):
        super(Hybrid3DCNN, self).__init__()
        self.base = video_models.r3d_18(pretrained=True)
        self.feature_dim = self.base.fc.in_features
        self.base.fc = nn.Identity()  # Remove classification layer
        self.classifier = nn.Linear(self.feature_dim, num_classes)

    def forward(self, x):
        features = self.base(x)
        logits = self.classifier(features)
        return logits, features
```

```python
from torch.utils.data import DataLoader
import torch

# Set device
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Dataset and DataLoader Setup
dataset = VideoClipDataset(root_dir="autism", clip_len=16, transform=transform)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True, num_workers=4)

# Define Model
model = Hybrid3DCNN(num_classes=5).to(device)  # Make sure num_classes=5 if using all 5 labels

from torch.optim import Adam
from torch.nn import CrossEntropyLoss

def train_model(model, dataloader, device, num_epochs=30, patience=5):
    optimizer = Adam(model.parameters(), lr=1e-4)
    criterion = CrossEntropyLoss()
    best_loss = float('inf')
    epochs_no_improve = 0

    model.to(device)

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0

        for batch_idx, (clips, labels) in enumerate(dataloader):
            clips, labels = clips.to(device), labels.to(device)
            optimizer.zero_grad()

            # Forward pass
            outputs, _ = model(clips)  # Or just outputs = model(clips) if your model returns one thing

            # Calculate loss
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
        print(f"Epoch {epoch+1} Loss: {running_loss:.4f}")

        # Early stopping
        if running_loss < best_loss:
            best_loss = running_loss
            torch.save(model.state_dict(), "best_model.pth")
            epochs_no_improve = 0
        else:
            epochs_no_improve += 1
            if epochs_no_improve >= patience:
                print("Early stopping triggered.")
                break
```

```python
def generate_confusion_matrix(model, dataloader, device):
    model.eval()
    all_preds = []
    all_labels = []

    # Collect predictions and true labels
    with torch.no_grad():
        for clips, labels in dataloader:
            clips, labels = clips.to(device), labels.to(device)
            outputs, _ = model(clips)
            _, preds = torch.max(outputs, 1)

            all_preds.extend(preds.cpu().numpy())  # move to CPU and store
            all_labels.extend(labels.cpu().numpy())  # move to CPU and store

    # Generate the confusion matrix
    cm = confusion_matrix(all_labels, all_preds)

    # Normalize the matrix to percentage (optional)
    cm_percentage = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    # Plot the confusion matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_percentage, annot=True, fmt=".2f", cmap="Blues", xticklabels=["not_present", "mild", "moderate", "marked"], yticklabels=["not_present", "mild", "moderate", "marked"])
    plt.title('Normalized Confusion Matrix')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

    return cm
# Assuming 'model' is your trained model and 'dataloader' is the DataLoader for your test/validation set
confusion_matrix_result = generate_confusion_matrix(model, dataloader, device)
```

```python
import torch
import torch.nn.functional as F
from torch.utils.data import DataLoader
from collections import defaultdict

def extract_average_embeddings(model, dataset_root, transform, device):
    class_embeddings = defaultdict(list)
    class_map = {'not_present': 0, 'mild': 1, 'moderate': 2, 'marked': 3}

    model.eval()

    with torch.no_grad():
        for class_name, class_idx in class_map.items():
            class_path = os.path.join(dataset_root, class_name)
            frame_files = sorted(glob.glob(os.path.join(class_path, "*.jpg")))
            for i in range(0, len(frame_files) - 15, 16):  # clip_len=16
                clip_frames = frame_files[i:i+16]
                clip = [transform(Image.open(f).convert("RGB")) for f in clip_frames]
                clip = torch.stack(clip, dim=1).unsqueeze(0).to(device)  # (1, C, T, H, W)

                _, features = model(clip)  # Assuming model returns (logits, features)
                class_embeddings[class_name].append(features.squeeze(0))

    avg_embeddings = {cls: torch.stack(embs).mean(dim=0) for cls, embs in class_embeddings.items()}
    return avg_embeddings
```

```python
def zero_shot_predict(model, test_clip, avg_embeddings, device):
    model.eval()
    with torch.no_grad():
        test_clip = test_clip.unsqueeze(0).to(device)  # (1, C, T, H, W)
        _, test_feat = model(test_clip)
        test_feat = test_feat.squeeze(0)

        similarities = {}
        for class_name, avg_feat in avg_embeddings.items():
            sim = F.cosine_similarity(test_feat, avg_feat, dim=0)
            similarities[class_name] = sim.item()

        # Determine if severe
        similarity_to_marked = similarities['marked']
        if all(sim >= similarity_to_marked for sim in similarities.values()):
            predicted_class = 'severe'
            confidence = 1.0
        else:
            predicted_class = max(similarities, key=similarities.get)
            confidence = similarities[predicted_class]

        return predicted_class, confidence
```

```python
import os
import cv2

def extract_frames_from_range(video_path, output_dir, start_time, end_time, fps=1):
    os.makedirs(output_dir, exist_ok=True)

    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"❌ Failed to open {video_path}")
        return

    video_fps = cap.get(cv2.CAP_PROP_FPS)
    start_frame = int(start_time * video_fps)
    end_frame = int(end_time * video_fps)

    cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

    frame_count = start_frame
    saved_count = 0
    interval = int(video_fps // fps)

    while frame_count <= end_frame:
        ret, frame = cap.read()
        if not ret:
            break

        if (frame_count - start_frame) % interval == 0:
            frame_filename = os.path.join(output_dir, f"frame_{saved_count:04d}.jpg")
            cv2.imwrite(frame_filename, frame)
            saved_count += 1

        frame_count += 1

    cap.release()
    print(f"✅ Extracted {saved_count} frames from {start_time}s to {end_time}s of {os.path.basename(video_path)}")

def extract_from_input():
    # Manually input video path
    video_path = input("Enter the full path to the video file (e.g., C:/videos/video.mp4): ").strip()

    if not os.path.exists(video_path):
        print(f"❌ Video path not found: {video_path}")
        return

    # Get start and end time input from user
    try:
        start_time = float(input("Enter start time in seconds: "))
        end_time = float(input("Enter end time in seconds: "))
        assert end_time > start_time, "End time must be greater than start time."
    except Exception as e:
        print(f"❌ Invalid input: {e}")
        return

    # Output folder (can be customized if you want to specify a different path)
    output_dir = os.path.join("test", "mild")
    os.makedirs(output_dir, exist_ok=True)

    # Extract frames
    extract_frames_from_range(video_path, output_dir, start_time, end_time, fps=1)

# Run the function
extract_from_input()
```
